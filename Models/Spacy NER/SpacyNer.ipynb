```python
 import spacy

 !python -m spacy download en_core_web_lg

 nlp = spacy.load("en_core_web_lg")
```

```python
import ast

def process_dataframe(df):

    df["entities"] = df['entities'].apply(ast.literal_eval)
    training_data = {'annotations': []}

    for index, row in df.iterrows():
        #print(row["entities"])
        temp_dict = {
            'text': row['text'],
            'entities': []
        }
        # Handling lists of lists in 'entities'

        for entity in list(row['entities']):
            #print(type(entity))
            #print(entity)
            start, end, label = entity  # Extracting from the list format
            # Convert label to uppercase and append as a tuple
            temp_dict['entities'].append((start, end, label.upper()))

        training_data['annotations'].append(temp_dict)

    return training_data
```

```python
testing_data = process_dataframe(test_data)
training_data = process_dataframe(train_data)
validating_data = process_dataframe(val_data)
```

```python
from spacy.util import filter_spans
from spacy.tokens import DocBin
from tqdm import tqdm

def convert_to_spacy_format(df):

    doc_bin = DocBin()

    for example  in tqdm(df['annotations']):
        text = example['text']
        labels = example['entities']
        doc = nlp.make_doc(text)
        ents = []
        for start, end, label in labels:
            span = doc.char_span(start, end, label=label, alignment_mode="contract")
            if span is None:
                print("Skipping entity")
            else:
                ents.append(span)
        filtered_ents = filter_spans(ents)
        doc.ents = filtered_ents
        doc_bin.add(doc)

    doc_bin.to_disk("validating_data.spacy")
```

```python
convert_to_spacy_format(training_data)
convert_to_spacy_format(testing_data)
convert_to_spacy_format(validating_data)
```

```python
base_config = """
# This is an auto-generated partial config. To use it with 'spacy train'
# you can run spacy init fill-config to auto-fill all other default settings.
[paths]
train = ./training_data.spacy
dev = ./validating_data.spacy

[system]
gpu_allocator = null

[nlp]
lang = "en"
pipeline = ["ner"]
batch_size = 1000

[training]
dropout = 0.2
batch_size = 1000
optimizer = {
  "@optimizers": "Adam.v1",
  "learn_rate": 0.001,
  "beta1": 0.9,
  "beta2": 0.999,
  "eps": 1e-08
}

[components]

[components.ner]
factory = "ner"
"""

# Write the base config to a file
with open('base_config.cfg', 'w') as file:
    file.write(base_config)
```
